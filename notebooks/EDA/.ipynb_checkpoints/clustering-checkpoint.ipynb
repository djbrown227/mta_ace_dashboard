```python
import os

os.getcwd()

from pathlib import Path

# Project root = two levels up from notebooks/EDA
PROJECT_ROOT = Path.cwd().parents[1]
```

```python
DATA_PROCESSED = PROJECT_ROOT / "data" / "processed"

DATA_PROCESSED.mkdir(parents=True, exist_ok=True)
```

```python
import numpy as np
import pandas as pd
processed_dfs = {
    csv_file.stem: pd.read_csv(csv_file)
    for csv_file in DATA_PROCESSED.glob("*.csv")
}
```

```python
for name, df in processed_dfs.items():
    print(f"\n===== {name} =====")
    print(df.head())
```

```python
renamed_dfs = {}

for name, df in processed_dfs.items():
    renamed_dfs[name] = (
        df[["route_id", "total_effect_pct"]]
        .rename(columns={"total_effect_pct": f"{name}_total_effect_pct"})
    )
```

```python
for name, df in renamed_dfs.items():
    print(df.head())
```

```python
from functools import reduce

df_all_effects = reduce(
    lambda left, right: left.merge(right, on="route_id", how="outer"),
    renamed_dfs.values()
)
```

```python
df_all_effects.head(5)
```

```python
df = df_all_effects.copy()
df = df[['route_id','speed_overall_total_effect_pct','journeyperformance_overall_total_effect_pct']]
df.head(5)

DATA_PROCESSED_1 = PROJECT_ROOT / "data"

DATA_PROCESSED_1.mkdir(parents=True, exist_ok=True)

pca_path = DATA_PROCESSED_1 / "df_speedjrny.csv"
df.to_csv(pca_path, index=False)
```

```python
cluster_features = [
    # raw outcomes
    "speed_overall_total_effect_pct",
    "journeyperformance_overall_total_effect_pct",

]

from sklearn.preprocessing import RobustScaler

X = df[cluster_features]
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
import pandas as pd

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(
    X_pca,
    columns=["PC1", "PC2"]
)
pca_df["route_id"] = df["route_id"]

```

```python
pca.explained_variance_ratio_
```

```python
from scipy.cluster.hierarchy import linkage

Z = linkage(X_scaled, method="ward")
```

```python
'''
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
dendrogram(
    Z,
    labels=df["route_id"].values,
    leaf_rotation=90
)
plt.tight_layout()
plt.show()
'''
```

```python
from scipy.cluster.hierarchy import fcluster

df["cluster_hc_4"] = fcluster(Z, t=4, criterion="maxclust")
```

```python
df["cluster_hc_2"] = fcluster(Z, t=2, criterion="maxclust")
df["cluster_hc_3"] = fcluster(Z, t=3, criterion="maxclust")
df["cluster_hc_5"] = fcluster(Z, t=5, criterion="maxclust")
```

```python
cluster_means = (
    df
    .groupby("cluster_hc_2")[cluster_features]
    .mean()
    .round(2)
)

cluster_means
```

```python
pca_df["cluster"] = df["cluster_hc_2"]

plt.figure()
for c in sorted(pca_df["cluster"].unique()):
    subset = pca_df[pca_df["cluster"] == c]
    plt.scatter(subset["PC1"], subset["PC2"], label=f"Cluster {c}")

plt.xlabel("PC1: Overall Effect")
plt.ylabel("PC2: Peak vs Reliability Skew")
plt.legend()
plt.show()
```

```python
pca_df = pca_df[['route_id','cluster']]
pca_df.head(5)

DATA_PROCESSED_1 = PROJECT_ROOT / "data"

DATA_PROCESSED_1.mkdir(parents=True, exist_ok=True)

pca_path = DATA_PROCESSED_1 / "pca_df.csv"
pca_df.to_csv(pca_path, index=False)
```

```python
# Select only routes in the non-improved cluster
df_non = df[df["cluster_hc_2"] == 2].copy()

# Examine speed vs reliability patterns
df_non["speed_only"] = (
    (df_non["speed_overall_total_effect_pct"] > 1) &
    (df_non["journeyperformance_overall_total_effect_pct"] < 0)
)

df_non["reliability_only"] = (
    (df_non["journeyperformance_overall_total_effect_pct"] > 1) &
    (df_non["speed_overall_total_effect_pct"] < 0)
)

df_non["both_positive"] = (
    (df_non["speed_overall_total_effect_pct"] > 0) &
    (df_non["journeyperformance_overall_total_effect_pct"] > 0)
)

df_non["both_negative_or_flat"] = (
    (df_non["speed_overall_total_effect_pct"] <= 0) &
    (df_non["journeyperformance_overall_total_effect_pct"] <= 0)
)

# Quick summary
#print(df_non[["route_id", "speed_only", "reliability_only", "both_positive", "both_negative_or_flat"]])
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Features for PCA
features_non = [
    "speed_overall_total_effect_pct",
    "speed_peak_total_effect_pct",
    "speed_offpeak_total_effect_pct",
    "journeyperformance_overall_total_effect_pct",
    "journeyperformance_peak_total_effect_pct",
    "journeyperformance_off_peak_total_effect_pct",
    "speed_peak_bias",
    "journey_peak_bias",
    "journey_vs_speed_gap"
]

X_non = df_non[features_non]

# Perform PCA
pca_non = PCA(n_components=2)
X_non_pca = pca_non.fit_transform(X_non)

# Plot PCA with mechanism categories
plt.figure(figsize=(8,5))

for idx in range(len(df_non)):
    row = df_non.iloc[idx]   # use iloc for correct row alignment
    x = X_non_pca[idx, 0]
    y = X_non_pca[idx, 1]
    
    # Assign color based on mechanism category
    if row["speed_only"]:
        color = "red"
        label = "Speed-only"
    elif row["reliability_only"]:
        color = "blue"
        label = "Reliability-only"
    elif row["both_positive"]:
        color = "green"
        label = "Both positive"
    else:
        color = "gray"
        label = "Negative/flat"
    
    plt.scatter(x, y, color=color, s=80)
    plt.text(x+0.1, y+0.1, row["route_id"], fontsize=8)

plt.xlabel("PC1: Overall effect magnitude")
plt.ylabel("PC2: Peak vs Reliability skew")
plt.title("Non-improved routes: mechanism decomposition")
plt.show()
```

```python
# Define a function for clear category labels
def categorize_route(row):
    if row["speed_only"]:
        return "Speed-only"
    elif row["reliability_only"]:
        return "Reliability-only"
    elif row["both_positive"]:
        return "Both positive"
    else:
        return "No improvement"

# Apply to non-improved cluster
df_non["mechanism_category"] = df_non.apply(categorize_route, axis=1)

# Quick summary of categories
print(df_non.groupby("mechanism_category")["route_id"].count())
```

```python
#This is for the boxplots showing Peak and OffPeak bus speed increases/decreases
# Converting to TypeScript
df = df_all_effects.copy()

df = df.rename(columns={
    "route_id": "routeId",
    "speed_peak_total_effect_pct": "peakEffectPct",
    "speed_offpeak_total_effect_pct": "offPeakEffectPct"
})

# Convert NaN  None
df = df.where(pd.notnull(df), None)


import math

def to_ts_value(v):
    if v is None:
        return "null"
    if isinstance(v, float) and math.isnan(v):
        return "null"
    return round(float(v), 4)


rows = []

for _, r in df.iterrows():
    rows.append(
        f"""  {{
    routeId: "{r.routeId}",
    peakEffectPct: {to_ts_value(r.peakEffectPct)},
    offPeakEffectPct: {to_ts_value(r.offPeakEffectPct)}
  }}"""
    )

#  FIX: join rows *outside* the f-string
rows_joined = ",\n".join(rows)

ts = f"""
export type SpeedPeakOffPeakRow = {{
  routeId: string;
  peakEffectPct: number | null;
  offPeakEffectPct: number | null
}};

export const speedPeakOffPeak: SpeedPeakOffPeakRow[] = [
{rows_joined}
];
"""

output_path = (
    "/Users/danielbrown/Desktop/mta-ace-buses/src/data/processed/speedPeakOffPeak.ts"
)

with open(output_path, "w") as f:
    f.write(ts)

print(" speedPeakOffPeak.ts written")
```
